```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import time

sns.set()
```

```python
def plot_avg_reward(experiment):
    df = pd.read_csv(f"experiments/{experiment}/progress.txt", delimiter="\t")
    sns_plot = sns.lineplot(data=df, x="Epoch", y="AverageEpRet")
#     sns_plot.savefig(f"{int(time.time())}.png")
```

```python
plot_avg_reward('elbert_clipnormalized')
```

```python
def plot_move_hist(moves):
    df = pd.read_csv(f"moves/{moves}.txt", delimiter="\t", header=None)
    sns.barplot(x=list(range(11)), y=df.mean(axis=0))
    
    plt.xlabel('Move index')
    plt.ylabel('Average frequency')
#     return df.mean(axis=0)
#     sns.distplot(df.mean(axis=0));
```
